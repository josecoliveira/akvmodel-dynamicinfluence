```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Example data
x = np.linspace(0, 10, 100)
y_data = [np.sin(x + i) for i in range(24)]
y2_data = [np.cos(x + i) for i in range(24)]
```

```python
fig, axes = plt.subplots(12, 4, figsize=(20, 30), sharex='col')

for i in range(24):
    row = i // 4
    col = i % 4
    
    # Plot the first graph of the pair
    axes[2*row, col].plot(x, y_data[i], label=f'Sin {i+1}')
    axes[2*row, col].legend()
    axes[2*row, col].set_title(f'Pair {i+1} - Graph 1')
    axes[2*row, col].grid(True)
    
    # Plot the second graph of the pair
    axes[2*row+1, col].plot(x, y2_data[i], label=f'Cos {i+1}')
    axes[2*row+1, col].legend()
    # axes[2*row+1, col].set_title(f'Pair {i+1} - Graph 2')
    axes[2*row+1, col].grid(True)

plt.tight_layout()
plt.show()
```
